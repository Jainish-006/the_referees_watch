```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolov8m.pt")  # load a pretrained model (recommended for training)

# Use the model

model.train(data="marks.yaml", epochs=100)
```

```python
metrics = model.val() 
```

```python
# predict on an image
success = model.export(format="pt")
```

```python
results = model("test9.mp4", save=True)  
```
